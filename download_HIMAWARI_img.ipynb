```python
import requests
import datetime
import os
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import time
```

```python
# Taiwan area 12 LST VIS image from Himawari
img_url = 'http://himawari8-dl.nict.go.jp/himawari8/img/D531106/20d/550/{:4d}/{:02d}/{:02d}/{:02d}0000_6_5.png' 
for yyyy in [2018, 2019, 2020]:
    for mm in range(1, 13):
        for dd in range(1, 32):
            # Check whether the date is valid
            isValidDate = True
            try :
                datetime.datetime(yyyy, mm, dd)
            except ValueError :
                isValidDate = False            
            if isValidDate and datetime.datetime(yyyy, mm, dd) < datetime.datetime(2020, 7, 13):
                # Check whether the image exists
                for zz in range(1, 9):
                    if not os.path.exists('img/vis_{:04d}_{:02d}_{:02d}_{:02d}LST.png'.format(yyyy, mm, dd, zz+8)):
                        # start download image
                        img_data = requests.get(img_url.format(yyyy, mm, dd, zz)).content
                        with open('img/vis_{:04d}_{:02d}_{:02d}_{:02d}LST.png'.format(yyyy, mm, dd, zz+8), 'wb') as handler:
                            handler.write(img_data)
                            print('Success download {}/{}/{} {:02d} VIS image!'.format(yyyy, mm, dd, zz+8))
                            time.sleep(5)
                    else:
                        print('{}/{}/{} {:02d} VIS image exists'.format(yyyy, mm, dd, zz+8))
                
```
